**Pre-Processing**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("data.csv")
data
```

```python
# Setting the decimal to 6 points
pd.options.display.float_format = '{:.6f}'.format
data.head()
```

```python
# Get info about the dataset
data.info()
```

```python
# Check if there are any null values in the dataset
data.isna().sum()
```

```python
# Count the number of outliers in each feature
def count_outliers(column):
    # Calculate the lower and upper bounds for the feature
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Find the indices of outliers
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers, lower_bound, upper_bound

for column in data.columns[:-1]:
    result = count_outliers(data[column])
    outliers = result[0]
    print("Number of outliers before handling in {0}: {1}".format(column, outliers.sum()))
```

```python
# Remove outliers in a column if number of outliers in column are less than 5% of total values in the column
for column in data.columns[:-1]:
    result = count_outliers(data[column])
    outliers = result[0]
    if (outliers.sum() <= 0.05 * len(data[column])):
        data.loc[outliers, column] = np.nan

# Remove rows with NaN values (rows that had outliers)
data = data.dropna()
data.shape
```

```python
# Count the number of outliers in each feature after 1st phase of outlier handling
for column in data.columns[:-1]:
    result = count_outliers(data[column])
    outliers = result[0]
    print("Number of outliers After 1st phase of handling in {0}: {1}".format(column, outliers.sum()))
```

```python
# Handle outliers such that if outlier value is close to lower_bound replace it with lower_bound else upper_bound
for column in data.columns[:-1]:
    result = count_outliers(data[column])
    outliers = result[0]
    lower_bound = result[1]
    upper_bound = result[2]
    data.loc[data[column] <= lower_bound, column] = lower_bound
    data.loc[data[column] >= upper_bound, column] = upper_bound

# Count the number of outliers in each feature after 2nd phase of outlier handling
for column in data.columns[:-1]:
    result = count_outliers(data[column])
    outliers = result[0]
    print("Number of outliers After 2nd phase of handling in {0}: {1}".format(column, outliers.sum()))
```

```python
# Normalization of the dataset

def normalization(column_name):
    min = data[column_name].min()
    max = data[column_name].max()
    data.loc[:, column_name] = (data[column_name] - min) / (max - min)
    
for feature in data.columns[:-1]:
    normalization(feature)

data.head()
```

```python
# Selecting columns with good correlation
selected_columns = []
correlation = data.corr()

for row in correlation.index:
    for col in correlation.columns:
        correlation_value = correlation.loc[row, col]
        if((correlation_value != 1) and (correlation_value >= 0.3) and (correlation_value <= 0.8)):
            if(col not in selected_columns):
                selected_columns.append(col)

# Sorting the columns
selected_columns = list(map(int,selected_columns))
selected_columns.sort()

# Selected columns
selected_columns = list(map(str,selected_columns))
print(selected_columns)
```

```python
# Deciding the features and labels
X = data[selected_columns]
y = data['48']
```

```python
# Split the data into training and test sets
x,y = X.sample(frac = 1, random_state = 42), y.sample(frac = 1, random_state = 42)
# Calculating 80% of data
x_train_size,y_train_size = int(x.shape[0]*0.80), int(y.shape[0]*0.80)

# Defining X_train,X_test, Y_train, Y_test
X_train,X_test = x[:x_train_size], x[x_train_size:]
Y_train,Y_test = y[:y_train_size], y[y_train_size:]

# Printing the spaes
print("Shape of X_train = {0} \nShape of X_test = {1} \nShape of Y_train = {2} \nShape of Y_test ={3}".format(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape))
```

**KNN Model**

```python
class KNN:
    def __init__(self):
        pass
        
    # Get the predicted values
    def fit(X_train, X_test, Y_train, k=3):
        Y_pred = []
        for sample in X_test:
            distances = np.sqrt(np.sum((X_train - sample) ** 2, axis=1))
            nearest_indices = np.argsort(distances)[:k]
            nearest_labels = Y_train[nearest_indices]
            unique_labels, counts = np.unique(nearest_labels, return_counts=True)
            prediction = unique_labels[np.argmax(counts)]
            Y_pred.append(prediction)
        Y_pred = np.array(Y_pred)
        return Y_pred

    def accuracy(self, y_true, y_pred):
        # Raise Exception for mismatch in lenghts of true and predicted values    
        if(len(y_true) != len(y_pred)):
            raise Exception("Lnegth of true and predicted value lists shall be the same")
        
        # Initialize correct predictions to 0
        correct_predictions = 0
        for i,j in zip(y_true, y_pred):
            if(i == j):
                correct_predictions += 1
                
        # Calculate the accuracy
        accuracy = correct_predictions / len(y_true)
        return accuracy

    def recall(self, y_true, y_pred):
        true_positive = sum((i == 1 and j == 1) for i,j in zip(y_true, y_pred))
        false_negative = sum((i == 1 and j == 0) for i,j in zip(y_true, y_pred))
        
        if(true_positive + false_negative == 0):
            return 0.0
    
        recall = true_positive / (true_positive + false_negative)
        return recall

    def precision(self, y_true, y_pred):
        true_positive = sum((i == 1 and j == 1) for i,j in zip(y_true, y_pred))
        false_positive = sum((i == 0 and j == 1) for i,j in zip(y_true, y_pred))
        
        if(true_positive + false_positive == 0):
            return 0.0
    
        recall = true_positive / (true_positive + false_positive)
        return recall

    def f1_score(self, y_true, y_pred):
        recall = self.recall(y_true,y_pred)
        precision = self.precision(y_true,y_pred)
        
        f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
        return f1_score 
```

```python
# Calculate the value of k

accuracy_history = []

for k in range(1, 20, 2):
    # Initialize the model
    model = KNN()
    Y_pred = KNN.fit(X_train.values, X_test.values, Y_train.values, k)
    # Calculate Evaluation metrices
    accuracy = model.accuracy(Y_test, Y_pred)
    recall = model.recall(Y_test, Y_pred)
    precision = model.precision(Y_test, Y_pred)
    f1_score = model.f1_score(Y_test, Y_pred)
    
    # Print the evaluation metrices
    print(f"k = {k}")
    print(f"Accuracy = {accuracy}")
    print(f"Recall = {recall}")
    print(f"Precision = {precision}")
    print(f"F1 Score = {f1_score}")
    print("\n")

    # Store accuracy history
    accuracy_history.append(accuracy)
```

```python
# Plot the elbow graph

plt.figure(figsize=(6, 4))
plt.plot(range(1,20,2), accuracy_history, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy Graph')
plt.grid(True)
plt.show()
```
